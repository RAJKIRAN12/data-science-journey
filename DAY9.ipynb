```python
# Group by one column and calculate the mean
import pandas as pd 

df = pd.read_csv('data.csv')

print(df.head())

grouped = df.groupby('Gender')['Salary'].mean()
print(grouped)

# Group by two columns
grouped2 = df.groupby(['Gender', 'Department'])['Salary'].mean()
print(grouped2)
```
